## Create new column for presence of any physical chronic condition

Heart Attack, Angina or Coronary Heart Disease, Stroke, Asthma, skin cancer, other types of cancer, chronic obstructive pulmonary disease, C.O.P.D., emphysema or chronic bronchitis, arthritis, rheumatoid arthritis, gout, lupus, or fibromyalgia, depressive disorder, kidney disease, diabetes

```python
#Lists of columns with chronic disease info
chronic_var_2020 = ['CVDINFR4', 'CVDCRHD4', 'CVDSTRK3', 'ASTHMA3', 'ASTHNOW', 'CHCSCNCR', 'CHCOCNCR', 'CHCCOPD2', 
                    'HAVARTH4', 'CHCKDNY2', 'DIABETE4']

#Add other years' here
```

```python
if year == 2020:
    current_year_chron = chronic_var_2020
#elif ....
```

```python
for column in current_year_chron:
    print(df[column].dtypes)
```

```python
def has_chronic_condition(row):
    for column in current_year_chron:
        if not isnan(row[column]) and round(row[column]) == 1: #Round for comparison to int
            return True
        
    return False
```

```python
df['CHRON_COND'] = df.apply(lambda row: has_chronic_condition(row), axis=1)
    
df.head(10)
```
